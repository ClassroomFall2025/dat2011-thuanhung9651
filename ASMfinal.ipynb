```python
from ASMfinal import QuanLyNhanVien, nhap_nv_tu_ban_phim, CSV_FILE
ql = QuanLyNhanVien()
print("Gọi các menu để bắt đầu chạy chương trình quản lý nhân viên")
def menu():
    while True:
        print("\n===== MENU =====")
        print("1. Nhập danh sách nhân viên và LƯU vào file")
        print("2. Đọc từ file & Xuất danh sách")
        print("3. Tìm nhân viên theo mã")
        print("4. Xóa nhân viên theo mã (ghi file)")
        print("5. Cập nhật thông tin nhân viên (ghi file)")
        print("6. Tìm theo khoảng lương")
        print("7. Sắp xếp theo họ tên")
        print("8. Sắp xếp theo thu nhập")
        print("9. Top 5 thu nhập cao nhất")
        print("0. Thoát")
        chon = input("Chọn: ")

        if chon == "1":
            n = int(input("Số lượng cần nhập: "))
            for _ in range(n):
                nv = nhap_nv_tu_ban_phim()
                if nv:
                    ql.them_nv(nv)
            ql.luu_file(CSV_FILE)

        elif chon == "2":
            ql.doc_file(CSV_FILE)
            ql.xuat_ds()

        elif chon == "3":
            ma = input("Nhập mã cần tìm: ")
            nv = ql.tim_theo_ma(ma)
            if nv: ql.xuat_ds([nv])
            else: print(" Không tìm thấy nhân viên!")

        elif chon == "4":
            ma = input("Nhập mã cần xóa: ")
            ql.xoa_theo_ma(ma, CSV_FILE)

        elif chon == "5":
            ma = input("Mã cần cập nhật: ")
            ten = input("Tên mới (Enter nếu không đổi): ")
            luong = input("Lương mới (Enter nếu không đổi): ")
            ds = input("Doanh số mới (TT): ")
            hh = input("Hoa hồng mới (TT): ")
            tn = input("Trách nhiệm mới (TP): ")
            ql.cap_nhat_va_ghi(
                ma,
                CSV_FILE,
                ho_ten=ten if ten else None,
                luong=float(luong) if luong else None,
                doanh_so=float(ds) if ds else None,
                hoa_hong=float(hh) if hh else None,
                trach_nhiem=float(tn) if tn else None
            )

        elif chon == "6":
            mn = float(input("Lương min: "))
            mx = float(input("Lương max: "))
            ql.xuat_ds(ql.tim_theo_khoang_luong(mn, mx))

        elif chon == "7":
            ql.sap_xep_theo_ten(); ql.xuat_ds()

        elif chon == "8":
            ql.sap_xep_theo_thu_nhap(True); ql.xuat_ds()

        elif chon == "9":
            ql.xuat_ds(ql.top5_thu_nhap())

        elif chon == "0":
            print(" Tạm biệt!")
            break
        else:
            print(" Lựa chọn không hợp lệ!")
menu()


```
